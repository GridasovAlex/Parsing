## Find()  Find_all()

```python
# нахожу один конкретный див
blog_name = soup.find('div', class_='blog__title').string
print(blog_name)

print()
# нахожу все заголовик пост тайтл
all_post_title = soup.find_all('div', class_='post__title')
for item in all_post_title:
    print(item.find('h3').text)

print()
#глубокий поиск
post_1_text = soup.find_all('div', class_='user__post__info')[0].find('div', class_="post__text").text
print(post_1_text.strip()) 

print()
post_2_text = soup.find_all('div', class_='user__post__info')[1].find('div', class_="post__text").text
print(post_2_text.strip()) 
```

## find_perent() find_perents()

```python
# возвращает блок до первого родителя
perent = soup.find_all('div', class_='post__title')[2].find_parent()
print(perent)

print("-----------------------------------------------------------------------------------------------------")
# Возвращает всех родителей
perents = soup.find('div', class_='post__title').find_parents()
print(perents)
```

## find_next() find_previous()

```python
next_elem = soup.find(class_="container").find_next()
print(next_elem.text)

previous_elem = soup.find_all('h1')[1].find_previous()
print(previous_elem) #если предыдущий элемент родитель, вернет целый блок
```

```python
game_name_list = []
game_genre_list = []
mean_estimation_list = []
game_platform_list = []

for n in range(1,6):
    url = f'https://www.playground.ru/games?release=all&sort=time&p={n}'
    igor = requests.get(url).text
    soup = BS(igor, 'lxml')

    items = soup.find_all('div', class_="info")
    for elem in items:
        # ищем имя
        name = elem.find('div', class_="media-heading title").find('a').text
        game_name_list.append(name)
            
        # ищем оценку
        # estimation = elem.find('span', class_="value js-game-rating-value").text
        # mean_estimation_list.append(estimation)
        
        # ищем жанры
        geners = elem.find('div', class_="meta-info").find_next().find_next_sibling().find_all('a')
        genre_list = []
        for gener in geners:
            genre_list.append(gener.get_text())
            genre_str = ", ".join(genre_list)
        game_genre_list.append(genre_str)

        # ищем платформы
        platforms = elem.find('div', class_="meta-info").find_next().find_next_sibling().find_next_sibling().find_all('span')
        platform_list = []
        for platform in platforms:
            platform_list.append(platform.get_text())
            platform_str = ", ".join(platform_list)
        game_platform_list.append(platform_str)

game_dt = pd.DataFrame({
    "Title":game_name_list,
    "Genres":game_genre_list,
    "Platforms":game_platform_list
})
display(game_dt.head())
```
